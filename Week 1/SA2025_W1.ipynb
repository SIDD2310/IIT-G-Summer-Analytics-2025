### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.


```python
# your code here
median_mpg = cars_df['mpg'].median()

above_median_mpg_cars = cars_df[cars_df['mpg'] > median_mpg]

above_median_mpg_cars['hp_per_weight'] = above_median_mpg_cars['horsepower'] / \
    above_median_mpg_cars['weight']

max_ratio = above_median_mpg_cars['hp_per_weight'].max()

best_cars = above_median_mpg_cars[above_median_mpg_cars['hp_per_weight'] == max_ratio]

print(best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin


```python
# your code here
grouped = cars_df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

pivot_table = grouped.pivot(index='model_year', columns='origin', values='mpg')

pivot_table.plot(marker='o', figsize=(10, 6))

plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.


```python
# your code here

import seaborn as sns

cars_df = cars_df.reset_index()

plt.figure(figsize=(10, 7))

sns.scatterplot(
    data=cars_df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(50, 300),
    alpha=0.7,
    palette='Set1'
)

# Titles and labels
plt.title(
    'Horsepower vs Weight of Cars\nColored by Origin, Sized by MPG', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin', loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).


```python
# your code here
cars_df = cars_df.reset_index()

year_col = 'model_year' if 'model_year' in cars_df.columns else 'year'

grouped = cars_df.groupby('name').agg(
    num_years=(year_col, 'nunique'),
    count=(year_col, 'count'),
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean')
).reset_index()

consistent_cars = grouped[
    (grouped['num_years'] >= 2) &
    (grouped['mpg_std'] < 1.0)
]

consistent_cars_sorted = consistent_cars.sort_values(
    by=['count', 'mpg_mean'],
    ascending=[False, False]
)

print("Consistent car models (std(mpg) < 1.0, ≥ 2 appearances):")
print(consistent_cars_sorted[['name', 'count', 'mpg_mean', 'mpg_std']])
```
